```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
  for i in range(0,k):
    for j in range(1,len(source)):
      source[j-1]=source[j]
    source[-1]=0
  return source

source=[10,20,30,40,50,60]
print(shift_left(source,3))




# Test 02: Rotate Left k cell
def rotate_left(source, k):
  for i in range(0,k):
    temp=source[0]
    for j in range(1,len(source)):
      source[j-1]=source[j]
    source[-1]=temp
  return source


source=[10,20,30,40,50,60]
print(rotate_left(source,3))



# Test 03: Shift Right k cell
def shift_right(source, k):
  for i in range(0,k):
    for j in range(len(source)-1,0,-1):
      source[j]=source[j-1]
    source[0]=0
  return source 

source=[10,20,30,40,50,60]
print(shift_right(source,3))




# Test 04: Rotate Right k cell
def rotate_right(source, k):
  for i in range(0,k):
    temp=source[-1]
    for j in range(len(source)-1,0,-1):
      source[j]=source[j-1]
    source[0]=temp
  return source 

source=[10,20,30,40,50,60]
print(rotate_right(source,3))



# Test 05: Remove an element from an array
def remove(source, idx):
  for j in range(idx+1,len(source)):
    source[j-1]=source[j]
  source[-1]=0
  return source

source=[10,20,30,40,50,0,0]
arr05=remove(source,2)
print(arr05)

# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
  temp=[0]*len(source)
  count=0
  for i in range(len(source)):
    if source[i] != element:
      temp[count]=source[i]
      count+=1
  source=temp

  return source

source=[10,2,30,2,50,2,2,0,0]
print(remove_all(source,2))


# Test 07: Splitting an Array
def split_array(a):
    left_sum=0
    for i in range(len(a)):
        left_sum += a[i]
        right_sum=0
        for j in range(i+1,len(a)):
            right_sum+= a[j]

        if left_sum==right_sum:
            return True
    return False

a=  [1, 1, 1, 2, 1] 
print(split_array(a))

a2 = [2, 1, 1, 2, 1] 
print(split_array(a2))

a3 = [10, 3, 1, 2, 10]
print(split_array(a3))


# Test 08: Max Bunch Count
def max_bunch(a):
    start=1 
    new_lst=[]
    count1=0
    for p in range(len(a)-1):
        for j in range(p,len(a)-1):
            if a[j]==a[j+1]:
                start+=1
            else:
                new_lst=new_lst+[start]
                start=1
    for i in range(len(new_lst)):
        for k in (i,len(new_lst)-1):
            if new_lst[k]>new_lst[i]:
                max= new_lst[k]
    return max


arr =  [1, 2, 2, 3, 4, 4, 4]
print(max_bunch(arr))

arr_2= [1,1,2, 2, 1, 1,1,1]
print(max_bunch(arr_2))
```
