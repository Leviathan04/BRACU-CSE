```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.

    self.head = None
    
    for i in range(len(a)) :
      node = Node(a[i],None,None)
      if i == 0 :
        self.head = node
        pre_node = node
      else :
        pre_node.next = node
        node.prev = pre_node
        pre_node = node
      pre_node.next = self.head
      self.head.prev = pre_node


  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To 
    start=self.head
    count=1
    while start.next!=self.head:
      start=start.next
      count+=1
    return(count)
    # print("-------count node-----------")
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do

    k=self.head
    while k.next!=self.head:
      print(k.element)
      k=k.next
      
      if k.next==self.head:
        print(k.element)
    # print('-------------------------------------')

  # prints the elements in the list backward
  def backwardprint(self):
    k = self.head
    while k.next != self.head :
      k = k.next
    end = k  
    while k.prev != end :
      print(k.element)
      k = k.prev 
      if k.prev == end :
        print(k.element)
        


  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    k = self.head
    count = 0
    obj = False
    last_idx = self.countNode() - 1 
    last_node = k.prev

    while k.next != self.head :
      if idx == last_idx :
        return last_node
      elif idx == count :
        return k
      else :
        k = k.next
        count += 1  
    if obj == False :
      print("index error")  
     
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    if elem == self.head.element:
      return 0
    else:
      temp = self.head.next
      count = 1
      while temp != self.head:
        if temp.element == elem:
          return count
        count += 1
        temp = temp.next
      return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    if idx < 0 or idx > self.countNode() :
      print("invalid index")
    else:
          
      last_idx = self.countNode() - 1 

      node = Node(elem,None,None)

      if idx == 0 :
        pre_node = self.head.prev
        post_node = self.head

        pre_node.next = node
        node.next = post_node

        post_node.prev = node
        node.prev = pre_node   

        self.head = node

      else :
        pre_node = self.nodeAt(idx-1)
        post_node = pre_node.next
          
        pre_node.next = node
        node.next = post_node

        post_node.prev = node
        node.prev = pre_node
        
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    if idx < 0 or idx > self.countNode() :
      print("invalid index")
    else :
          
      last_idx = self.countNode() - 1 

      if idx == 0 :              
        r = self.head
        pre_node = self.head.prev
        post_node = self.head.next

        pre_node.next = post_node

        post_node.prev = pre_node   

        self.head = post_node
        return str(r.element)    
      else:
        r = self.nodeAt(idx)
        pre_node = r.prev
        post_node = r.next
          
        pre_node.next = post_node

        post_node.prev = pre_node   
        return str(r.element)
```
