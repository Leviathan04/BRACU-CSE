```python
#Task 2
class Assassin:
  count=0
  def __init__(self,name,rate):
    self.name = name
    self.rate = rate
    type(self).count+=1

  @classmethod
  def failureRate(cls,name2,rate2):
    rate = 100-rate2

    return cls(name2,rate)
  @classmethod
  def failurePercentage(cls,name3,rate3):
    rate = 100-int(rate3.strip("%"))
    return Assassin(name3,rate)
  def printDetails(self):
    print(f'Name: {self.name}\nSuccess rate: {self.rate}%\nTotal number of Assassin: {self.count}')




# Write your code here
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', "10%")
akabane.printDetails()
```

```python
#Task 3
class Passenger:
  count=0
  def __init__(self,name):
    self.name=name
    self.fare=450
    type(self).count+=1

  def set_bag_weight(self,weight):
    if weight <=20:
      self.fare ==0
    elif 21<=weight <=50:
      self.fare +=50
    elif 50<weight:
      self.fare +=100

  def printDetail(self):
    print(f'Name: {self.name}\nBus Fare: {self.fare} taka')



# Write your code here
print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)
```

```python
#Task 7
class Cat:
  Number_of_cats=0
  def __init__(self,color="White",status="sitting" ):
    self.color=color
    self.status=status
    Cat.Number_of_cats += 1
  @classmethod
  def no_parameter(cls):
    color='White'
    status = "sitting"
    return cls(color,status)
  @classmethod
  def first_parameter(cls,color2):
    color=color2
    return cls(color)
  @classmethod
  def second_parameter(cls,status2):
    status=status2
    color="Grey"
    return cls(color,status)
  def changeColor(self,color2):
    self.color=color2

  def printCat(self):
    print(f'{self.color} cat is {self.status}')







# Write your code here
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8
import math
class Cylinder:
  radius = 5
  height=18
  def __init__(self,radius,height):
    self.radius=radius
    self.height=height
    print(f'Default radius={Cylinder.radius} and height={Cylinder.height}.')
    Cylinder.radius=self.radius 
    Cylinder.height=self.height

    print(f'Updated: radius={Cylinder.radius} and height={Cylinder.height}.')

  @classmethod
  def swap(cls,radius,height):
    return cls(height,radius)
  @classmethod
  def changeFormat(cls,arg):
    data=arg.split('-')
    radius= float(data[0])
    height= float(data[1])
    return cls(radius,height)

  @staticmethod
  def area(radius,height):
    area = (2*math.pi*(radius**2))+(2*math.pi*radius*height)
    print(f"Area: {area}")
  @staticmethod
  def volume(radius,height):
    volume=math.pi*(radius**2)*height
    print(f'Volume: {volume}')
  



# Write your code here
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10
class SultansDine:
  branch=0
  total=0
  info=[]
  def __init__(self,place):
    self.place=place
    type(self).branch+=1

  def sellQuantity(self,quantity):
    if quantity<10:
      self.branch_sell =quantity*300
    elif quantity<20:
      self.branch_sell = quantity*350
    else:
      self.branch_sell =quantity*400
    type(self).total+=self.branch_sell

  def branchInformation(self):
    print(f'Branch Name: {self.place}\nBranch Sell: {self.branch_sell} Taka')
    type(self).info.append([self.place,self.branch_sell])

  @classmethod
  def details(cls):
    print(f"Total number of branch(s): {cls.branch}\nTotal sell: {cls.total} Taka")
    for i in range(len(cls.info)):
      percent = (cls.info[i][1] / cls.total) * 100
      print(f"Branch Name: {cls.info[i][0]}, Branch Sell: {cls.info[i][1]} Taka\nBranch consists of total sell's: {percent:.2f}%")



# Write your code here
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
